```python
# Hiệp phương sai cho từng sản phẩm
# Lấy danh sách các dòng sản phẩm  
product_lines = dataBLK['Dòng sản phẩm'].unique()  

# Tạo một từ điển để lưu kết quả  
covariance_results = {}  

# Duyệt qua từng dòng sản phẩm  
for product in product_lines:  
    # Lọc dữ liệu theo dòng sản phẩm  
    product_data = dataBLK[dataBLK['Dòng sản phẩm'] == product]  
    
    # Tính hiệp phương sai giữa các cặp biến  
    cov_price_quantity = product_data[['Giá', 'Số lượng']].corr().iloc[0, 1]  
    cov_price_revenue = product_data[['Giá', 'Doanh thu']].corr().iloc[0, 1]  
    cov_quantity_revenue = product_data[['Số lượng', 'Doanh thu']].corr().iloc[0, 1]  
    
    # Lưu kết quả vào từ điển  
    covariance_results[product] = {  
        'hiệp phương sai (giá, số lượng)': cov_price_quantity,  
        'hiệp phương sai (giá, doanh thu)': cov_price_revenue,  
        'hiệp phương sai (số lượng, doanh thu)': cov_quantity_revenue  
    }  

# Chuyển đổi kết quả thành DataFrame để dễ đọc  
covariance_df = pd.DataFrame(covariance_results).T  

# In kết quả  
print(covariance_df)  
```

## 10.1.Tìm giá tối ưu 1 ngày tiếp theo

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR
from scipy.optimize import fsolve

# Hàm cầu từ mô hình VAR
def demand(price, alpha, beta_price, beta_quantity):
    return alpha + np.dot(beta_price, price) + np.dot(beta_quantity, quantity)  # Cập nhật demand cho phù hợp

# Hàm doanh thu (Revenue)
def revenue(price,target_revenue):
    quantity= SV[0]+SV[1]*pr[1065]+(-SV[1]+SV[3])*pr[1064]+(-SV[3]+SV[5])*pr[1063]+(-SV[5]+SV[7])*pr[1062]+(-SV[7]+SV[9])*pr[1061]+(-SV[9]+SV[11])*pr[1060]-SV[11]*pr[1059]+(SV[2]+1)*vl[1065]+(-SV[2]+SV[4])*vl[1064]+(-SV[4]+SV[6])*vl[1063]+(-SV[6]+SV[8])*vl[1062]+(-SV[8]+SV[10])*vl[1061]+(-SV[10]+SV[12])*vl[1060]-SV[12]*vl[1059]
    return -price * quantity + target_revenue

# Tạo vòng lặp để giải mức giá tối ưu cho từng ngày với doanh thu cho trước
target_revenue = DT_yeucau[0]  # Mức doanh thu cần đạt
optimal_price = fsolve(revenue,1, args=( target_revenue))
optimal_price
```

## 10.2.Tìm giá tối ưu cho 30 ngày tiếp theo

```python
from scipy.optimize import minimize
import numpy as np

#Hàm cầu:
def demand(P_lags,Q_lags,P_t):
    const = SV[0]
    beta = np.array([SV[1],(-SV[1]+SV[3]),(-SV[3]+SV[5]),(-SV[5]+SV[7]),(-SV[7]+SV[9]),(-SV[9]+SV[11]),(-SV[11])])
    gamma = np.array([(SV[2]+1),(-SV[2]+SV[4]),(-SV[4]+SV[6]),(-SV[6]+SV[8]),(-SV[8]+SV[10]),(-SV[10]+SV[12]),(-SV[12])])
    #Tính Q_t
    Q_t = const +np.dot(beta,P_lags)+np.dot(gamma,Q_lags)+0*P_t
    return Q_t

#Hàm doanh thu
def revenue(P_t,P_lags,Q_lags,R_target):
    Q_t=demand(P_lags,Q_lags,P_t)
    return (P_t*Q_t-R_target)**2

```

```python
P_lags= np.array(pr[1059:])
Q_lags = np.array(vl[1059:])
P_lags = P_lags[::-1]
Q_lags = Q_lags[::-1]
```

```python
#Doanh thu mục tiêu trong 30 ngày tới
R_targets = DT_yeucau

#Tạo bảng lưu kết quả
optimal_prices = []
optimal_quantities = []

#Vòng lặp
for day in range(30):
    result = minimize(lambda P_t:revenue(P_t,P_lags,Q_lags,R_targets[day]),x0=50)
    P_t = result.x[0]
    #Tính lượng cầu ngày hiện tại
    Q_t = demand(P_lags,Q_lags,P_t)
    
    #Lưu kết quả
    optimal_prices.append(P_t)
    optimal_quantities.append(Q_t)
    
    #Cập nhật đỗ trễ cho ngày tiếp theo
    P_lags =np.roll(P_lags,-1)
    P_lags[-1]=P_t
    Q_lags = np.roll(Q_lags,-1)
    Q_lags[-1]=Q_t
    
for day in range(30):
    print(f"Ngày {day+1}:Mức giá chung tối ưu = {optimal_prices[day]:.2f},Số lượng = {optimal_quantities[day]:2f}")
    print("_____________________________________________________________")
    
```
